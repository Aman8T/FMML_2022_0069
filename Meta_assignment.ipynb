# Preprocessing and Classification


```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, valid_index in split.split(df, df['CLASS']):
    X_train = df.iloc[train_index]
    X_valid = df.iloc[valid_index]
```

```python
X_train['CLASS'].value_counts() / len(X_train)
```

```python
X_valid['CLASS'].value_counts() / len(X_valid)
```

```python
y = X_train['CLASS']
y_valid = X_valid['CLASS']
X_train.pop('CLASS')
X_valid.pop('CLASS')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_classif
```

```python
scaler=StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train)
X_valid = scaler.fit_transform(X_valid)
X_valid = pd.DataFrame(X_valid)
```

```python
import tensorflow 
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras import optimizers
import math
import keras.callbacks
```

Basic CNN Model

```python
#BASE MODEL
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=X_train.shape[1:]))
model.add(keras.layers.Dense(50,
                                 kernel_initializer="lecun_normal",
                                 activation="selu"))
model.add(keras.layers.Dense(50,
                                 kernel_initializer="lecun_normal",
                                 activation="selu"))

model.add(keras.layers.Dense(1, activation="sigmoid"))

optimizer=keras.optimizers.SGD(learning_rate=1e-2, momentum=0.9)
model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
```

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10,
                                                  restore_best_weights=True)
#onecycle = OneCycleScheduler(math.ceil(len(X) / 32) * 32, max_rate=0.05)


history=model.fit(X_train, y, epochs=100,
          validation_data=(X_valid,y_valid),
          callbacks = early_stopping_cb, batch_size=32)# class_weight=class_weights)
```

```python
X_train.shape
```

```python
y.shape
```

```python
from sklearn.metrics import classification_report
import numpy as np
```

```python
predictions = model.predict(X_valid)
predictions=np.argmax(predictions, axis=1)

print(classification_report(y_valid, predictions))
```

```python
model.evaluate(X_valid, y_valid)
```

```python
pca = PCA(n_components=3)
pca_fit = pca.fit_transform(a)
```

```python
print(pca_fit)
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
final_df = pd.DataFrame(columns=["PC1", "PC2", "Label"])

for i in range(len(pca_fit)):
    
    dicti = dict()
    
    dicti["PC1"] = pca_fit[i, 0]
    dicti["PC2"] = pca_fit[i, 1]
    dicti["PC3"] = pca_fit[i, 2]
    
    if  df['CLASS'][i] == 0:
      dicti["Label"] = "Normal"
    else:
      dicti["Label"] = "A"
    
    final_df = final_df.append(dicti, ignore_index = True)

final_df.head()
```

```python
final_df.tail()
```

```python
for_x = final_df.PC1.tolist()
for_y = final_df.PC2.tolist()
for_label = final_df.Label.tolist()
for_hover = final_df.Label.tolist()

fig = px.scatter(x=for_x, y=for_y, color=for_label,
                 title="Principal Component Axis",
                 color_discrete_map={"Normal": "aqua", "A": "yellow"})

fig.update_layout(
    xaxis=dict(title = 'PC1', showgrid=True, ticks='inside', zeroline=True, mirror=True, showline=True, linecolor='white'),
    yaxis=dict(title = 'PC2', showgrid=True, ticks='inside', zeroline=True, mirror=True, showline=True, linecolor='white'),
    plot_bgcolor='#555555',
    font=dict(
        family="Times New Roman",
        size=16,
        color="Black"))
fig.update_traces(marker=dict(size=8,))

fig.show(renderer = "colab")
```
